```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import joblib

# Load Dataset
df = pd.read_csv("/content/archive (5).zip")

# Features and Target
X = df[['TV', 'Radio', 'Newspaper']]
y = df['Sales']

# Train Model
model = LinearRegression()
model.fit(X, y)

# Save Model
joblib.dump(model, "sales_model.pkl")

print("Model Saved Successfully")

# Load Saved Model
model = joblib.load("sales_model.pkl")

# User Input Values
tv = 230.1
radio = 37.8
newspaper = 69.2

# Create DataFrame
input_data = pd.DataFrame({
    'TV': [tv],
    'Radio': [radio],
    'Newspaper': [newspaper]
})

# Predict Sales
prediction = model.predict(input_data)

print("\nSales Prediction App")
print("--------------------")
print("TV Advertising Budget:", tv)
print("Radio Advertising Budget:", radio)
print("Newspaper Advertising Budget:", newspaper)
print(f"\nPredicted Sales = {prediction[0]:.2f}")
```
